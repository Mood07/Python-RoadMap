## Section 1 — Comparison Operators

Comparison operators compare two values and always return a `bool` (`True`/`False`).

| Operator | Meaning |
|----------|---------|
| `==` | equal to |
| `!=` | not equal to |
| `<` | less than |
| `>` | greater than |
| `<=` | less than or equal to |
| `>=` | greater than or equal to |

> **Chained comparisons:** `0 < x < 100` is equivalent to `(0 < x) and (x < 100)` — unique to Python, very readable.

````python
x = 10
y = 3

print(x == 10)              # True
print(x != y)               # True
print(x > y)                # True
print(x < y)                # False
print(x >= 10)              # True
print(x <= 9)               # False

# Python compares int and float by value — no type mismatch
print(3 == 3.0)             # True   (same numeric value)
print(3 != 3.0)             # False

# Strings: lexicographic (alphabetical) order
print("apple" < "banana")   # True  ('a' comes before 'b')
print("Z" < "a")            # True  (uppercase codes are smaller)

# Chained comparisons — Pythonic range checks
age = 25
print(18 <= age < 65)       # True   (adult working age)
print(0 < age < 18)         # False  (not a minor)

score = 75
print(60 <= score <= 79)    # True   (C grade range)
````

## Section 2 — Logical Operators

Combine boolean conditions with `and`, `or`, `not`.

| Operator | Rule |
|----------|------|
| `and` | `True` only if **both** sides are truthy |
| `or` | `True` if **at least one** side is truthy |
| `not` | Inverts a boolean — always returns `True` or `False` |

**Short-circuit evaluation:**
- `and` stops at the **first falsy** value and returns it.
- `or` stops at the **first truthy** value and returns it.
- Neither evaluates the right side if it's not needed.

```python
# Default value pattern
display = username or "Anonymous"
```

````python
has_ticket = True
is_adult   = True
is_vip     = False

print(has_ticket and is_adult)  # True   (both True)
print(has_ticket and is_vip)    # False  (one False)
print(is_vip or is_adult)       # True   (one True)
print(not is_vip)               # True   (inverts False)
print(not has_ticket)           # False  (inverts True)

# Short-circuit: and/or return the deciding operand — not always a bool
print(0 and "hello")    # 0       (0 is falsy → return it, skip "hello")
print(5 and "hello")    # hello   (5 is truthy → evaluate and return "hello")
print(0 or "hello")     # hello   (0 is falsy → check next → "hello" is truthy)
print("hi" or "hello")  # hi      ("hi" is truthy → return it, skip "hello")
print(False or 0 or "") #         (all falsy → returns last value "")

# Practical: default value when a variable might be empty
username = ""
display  = username or "Anonymous"
print(display)          # Anonymous

username = "Berke"
display  = username or "Anonymous"
print(display)          # Berke

# not always returns a bool
print(not 0)            # True
print(not "hello")      # False
print(type(not 0))      # <class 'bool'>
````

## Section 3 — Identity Operators

| Operator | Tests |
|----------|-------|
| `is` | Same object in memory (identity) |
| `is not` | Different objects in memory |

`==` compares **values**. `is` compares **identity** (memory address).

> **Rule:** Use `is` only for `None`, `True`, and `False` — they are singletons.  
> Never use `is` for numbers or strings in general; Python caches small values but this is an implementation detail.

````python
a = [1, 2, 3]
b = [1, 2, 3]
c = a               # c points to the same list object as a

print(a == b)       # True   (same content)
print(a is b)       # False  (different objects in memory)
print(a is c)       # True   (same object — c is just another name for a)
print(a is not b)   # True

# Correct use of is: checking against None, True, False
value = None
print(value is None)     # True  ← correct idiom
print(value == None)     # True  (works, but PEP 8 says use 'is')

# Do NOT use is for arbitrary numbers
x = 1000
y = 1000
print(x == y)       # True   (always correct — use this)
print(x is y)       # False  (may vary — don't rely on is for numbers)
````

## Section 4 — Membership Operators

| Operator | Tests |
|----------|-------|
| `in` | Left value **found** in the right collection |
| `not in` | Left value **not found** in the right collection |

Works on: `str` (substring check), `list`, `tuple`, `set`, `dict` (checks **keys**).

````python
# Strings: substring check
text = "Hello, Python!"
print("Python" in text)     # True
print("Java" in text)       # False
print("java" not in text)   # True   (case-sensitive)

# Lists
primes = [2, 3, 5, 7, 11]
print(5 in primes)          # True
print(4 in primes)          # False
print(4 not in primes)      # True

# Tuples
vowels = ("a", "e", "i", "o", "u")
print("e" in vowels)        # True

# Dicts: checks KEYS, not values
scores = {"Alice": 95, "Bob": 80}
print("Alice" in scores)        # True   (checks keys)
print(95 in scores)             # False  (95 is a value, not a key)
print(95 in scores.values())    # True   (check values explicitly)
````

## Section 5 — Bitwise Operators

Bitwise operators work on the **binary representation** of integers.

| Operator | Name | Rule |
|----------|------|------|
| `&` | AND | 1 only where **both** bits are 1 |
| `\|` | OR | 1 where **at least one** bit is 1 |
| `^` | XOR | 1 only where bits **differ** |
| `~` | NOT | Flips every bit — result = `-(n + 1)` |
| `<<` | Left shift | Shift bits left (multiply by 2 per step) |
| `>>` | Right shift | Shift bits right (floor-divide by 2 per step) |

**Permission flag pattern** — each flag is a distinct power of 2:

```python
READ = 1   WRITE = 2   EXECUTE = 4
perms |= FLAG       # grant
perms &= ~FLAG      # revoke
bool(perms & FLAG)  # check
```

````python
a = 0b1010      # decimal 10  (binary: 1010)
b = 0b1100      # decimal 12  (binary: 1100)

print(bin(a & b))   # 0b1000  (8)  — only bit 3 is 1 in both
print(bin(a | b))   # 0b1110  (14) — bits set in either
print(bin(a ^ b))   # 0b0110  (6)  — bits that differ
print(~a)           # -11     — flips all bits: -(10 + 1)

print(a << 1)       # 20  — shift left 1  = multiply by 2
print(a << 2)       # 40  — shift left 2  = multiply by 4
print(a >> 1)       # 5   — shift right 1 = floor-divide by 2

# Permission flag system
READ    = 0b001     # 1
WRITE   = 0b010     # 2
EXECUTE = 0b100     # 4

user_perms = READ | WRITE           # 0b011 = 3 (has read + write)

print(bool(user_perms & READ))      # True   (can read)
print(bool(user_perms & WRITE))     # True   (can write)
print(bool(user_perms & EXECUTE))   # False  (cannot execute)

user_perms |= EXECUTE               # grant execute
print(bool(user_perms & EXECUTE))   # True   (now can execute)

user_perms &= ~WRITE                # revoke write
print(bool(user_perms & WRITE))     # False  (write revoked)
````

## Section 6 — Augmented Assignment Operators

All augmented operators update a variable using itself.

| Category | Operators |
|----------|-----------|
| Arithmetic | `+=` `-=` `*=` `/=` `//=` `%=` `**=` |
| Bitwise | `&=` `\|=` `^=` `<<=` `>>=` |

````python
# Bitwise augmented assignment
flags = 0b0000
flags |= 0b0001     # set bit 0  →  0b0001
flags |= 0b0100     # set bit 2  →  0b0101
print(bin(flags))   # 0b101

flags &= ~0b0001    # clear bit 0 →  0b0100
print(bin(flags))   # 0b100

flags ^= 0b1111     # toggle all 4 bits  →  0b1011
print(bin(flags))   # 0b1011

flags <<= 2         # shift left 2  →  0b101100
print(bin(flags))   # 0b101100

flags >>= 3         # shift right 3 →  0b101
print(bin(flags))   # 0b101

# Arithmetic augmented assignment
n = 20
n +=  5 ;  print(n)    # 25   (n = n + 5)
n -=  3 ;  print(n)    # 22   (n = n - 3)
n *=  2 ;  print(n)    # 44   (n = n * 2)
n //= 3 ;  print(n)    # 14   (n = n // 3)
n %=  4 ;  print(n)    # 2    (n = n % 4)
n **= 3 ;  print(n)    # 8    (n = n ** 3)
````

## Section 7 — Ternary / Conditional Expression

A one-line `if/else` that evaluates to a value.

```python
value_if_true if condition else value_if_false
```

Use when the expression is short and the intent is clear. Avoid nesting more than one level deep.

````python
score = 72
result = "Pass" if score >= 60 else "Fail"
print(result)           # Pass

age = 17
label = "Adult" if age >= 18 else "Minor"
print(label)            # Minor

# Works inside f-strings
speed = 120
status_msg = f"Speed: {speed} — {'OK' if speed <= 110 else 'TOO FAST'}"
print(status_msg)       # Speed: 120 — TOO FAST

# Absolute value without abs()
x = -42
abs_x = x if x >= 0 else -x
print(abs_x)            # 42

# Nested ternary (use sparingly)
score = 85
grade = "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "F"
print(grade)            # B

# Default when value might be None
value = None
safe = value if value is not None else "default"
print(safe)             # default

value = 42
safe = value if value is not None else "default"
print(safe)             # 42
````

## Section 8 — Operator Precedence

When multiple operators appear in one expression, Python evaluates them in this order (highest → lowest priority):

| Priority | Operators |
|----------|-----------|
| 1 (highest) | `()` — Parentheses |
| 2 | `**` — Exponentiation (right-to-left) |
| 3 | `+x` `-x` `~x` — Unary |
| 4 | `*` `/` `//` `%` — Multiply / divide |
| 5 | `+` `-` — Add / subtract |
| 6 | `<<` `>>` — Bitwise shifts |
| 7 | `&` — Bitwise AND |
| 8 | `^` — Bitwise XOR |
| 9 | `\|` — Bitwise OR |
| 10 | `==` `!=` `<` `>` `<=` `>=` `is` `is not` `in` `not in` |
| 11 | `not` — Logical NOT |
| 12 | `and` — Logical AND |
| 13 (lowest) | `or` — Logical OR |

> **Rule of thumb:** when in doubt, use parentheses — they always override.

````python
print(2 + 3 * 4)                    # 14   (* before +)
print(True or False and False)      # True  (and before or)
print(not True or False)            # False ((not True)=False, False or False)
print(not (True or False))          # False (parens first: True, then not)

print(5 > 3 and 10 < 20)           # True  (comparisons before and)
print(1 + 1 == 2)                   # True  (+ before ==)

# Chained comparisons
x = 5
print(1 < x < 10)                   # True   (Pythonic range check)
print(1 < x and x < 10)            # True   (same, more verbose)

# Parentheses change evaluation order
print((2 + 3) * 4)                  # 20    (parens applied first)
print(not (5 > 3 and 10 < 5))      # True
````

## Quick Reference Cheatsheet

```python
# Comparison — always returns bool
==  !=  <  >  <=  >=
0 < x < 100            # chained range check

# Logical
and   or   not
# Short-circuit: and → first falsy; or → first truthy
result = value or "fallback"   # default value pattern

# Identity — use ONLY for None, True, False
x is None
x is not None

# Membership
x in collection
x not in collection
# Works on: str (substring), list, tuple, set, dict (keys)

# Bitwise
&   AND   — both bits must be 1
|   OR    — at least one bit is 1
^   XOR   — bits must differ
~   NOT   — -(n+1)
<<  left shift  = multiply by 2^n
>>  right shift = floor-divide by 2^n
# Permission pattern:
perms |= FLAG         # grant
perms &= ~FLAG        # revoke
bool(perms & FLAG)    # check

# Augmented assignment
+=  -=  *=  /=  //=  %=  **=  &=  |=  ^=  <<=  >>= 

# Ternary
value_if_true if condition else value_if_false

# Precedence (high → low)
()  **  unary  * / // %  + -  << >>  &  ^  |  comparisons  not  and  or
```

> ⚠️ **Common traps:**
> - `is` compares identity, not value — `x is 1000` may be `False` even when `x == 1000`
> - `not` always returns `bool`; `and`/`or` return the deciding operand (may not be `bool`)
> - `and` binds tighter than `or` — `A or B and C` means `A or (B and C)`
