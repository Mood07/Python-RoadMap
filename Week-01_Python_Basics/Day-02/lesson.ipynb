## Section 1 — String Creation

A **string** is an ordered sequence of characters. Single and double quotes are interchangeable. Triple quotes let you write multi-line strings.

````python
single = 'Python'
double = "Python"
print(single == double)         # True — same content

sentence = "It's a beautiful day."
dialogue = 'She said, "Hello!"'

haiku = """Old pond,
a frog jumps in —
sound of water."""

print(sentence)
print(dialogue)
print(haiku)
````

## Section 2 — String Indexing

Every character has a **position** (index). Python supports positive (left-to-right) and negative (right-to-left) indexing.

```
String:    P  y  t  h  o  n
Positive:  0  1  2  3  4  5
Negative: -6 -5 -4 -3 -2 -1
```

````python
word = "Python"

print(word[0])                  # P  — first character
print(word[5])                  # n  — last character
print(word[-1])                 # n  — last (negative)
print(word[-6])                 # P  — first (negative)

# Uncomment to see the error:
# print(word[10])               # IndexError: string index out of range
````

## Section 3 — String Slicing

Slice syntax: **`string[start:stop:step]`**

| Parameter | Meaning | Default |
|---|---|---|
| `start` | Index to begin at (inclusive) | `0` |
| `stop` | Index to stop before (exclusive) | `len(string)` |
| `step` | How many positions to jump | `1` |

````python
text = "Hello, World!"

print(text[0:5])                # Hello
print(text[7:12])               # World
print(text[:5])                 # Hello   (start defaults to 0)
print(text[7:])                 # World!  (stop defaults to end)
print(text[:])                  # Hello, World!  (full copy)
print(text[::2])                # Hlo ol! (every 2nd character)
print(text[::-1])               # !dlroW ,olleH  (reversed)
print(text[-6:-1])              # World   (negative indices in slice)
````

## Section 4 — Escape Characters

| Escape | Meaning |
|---|---|
| `\n` | Newline |
| `\t` | Tab |
| `\\` | Literal backslash |
| `\'` | Literal single quote |
| `\"` | Literal double quote |

Prefix a string with **`r`** to make it a **raw string** — backslashes are treated literally.

````python
path = "C:\\Users\\Berke\\Documents"
print(path)                     # C:\Users\Berke\Documents

tabbed = "Name:\tBerke"
print(tabbed)                   # Name:    Berke

multiline = "Line 1\nLine 2\nLine 3"
print(multiline)

raw_path = r"C:\Users\Berke\new_folder"
print(raw_path)                 # C:\Users\Berke\new_folder
````

## Section 5 — String Methods: Transforming Text

> **Key rule:** Strings are **immutable**. Methods always return a **new** string — the original is never changed.

| Method | What it does |
|---|---|
| `.upper()` | All uppercase |
| `.lower()` | All lowercase |
| `.title()` | First letter of each word capitalised |
| `.capitalize()` | First letter of the whole string capitalised |
| `.strip()` | Remove leading/trailing whitespace |
| `.replace(old, new)` | Swap every occurrence of `old` with `new` |
| `.split(sep)` | Break into a list at `sep` |
| `sep.join(list)` | Glue a list together with `sep` |

````python
original = "  hello, world!  "
print(original.strip())         # hello, world!
print(original.lstrip())        # hello, world!   (left only)
print(original.rstrip())        # hello, world!   (right only)

phrase = "hello, world!"
print(phrase.upper())           # HELLO, WORLD!
print(phrase.lower())           # hello, world!
print(phrase.title())           # Hello, World!
print(phrase.capitalize())      # Hello, world!

sentence = "Python is great. Python is fun."
print(sentence.replace("Python", "Coding"))

csv_line = "Berke,25,developer"
parts = csv_line.split(",")
print(parts)                    # ['Berke', '25', 'developer']

words = ["Python", "is", "powerful"]
print(" ".join(words))          # Python is powerful
print("-".join(words))          # Python-is-powerful
````

## Section 6 — String Methods: Searching Text

| Method / Operator | Returns | Not found |
|---|---|---|
| `.find(sub)` | Index of first match | `-1` |
| `.index(sub)` | Index of first match | `ValueError` |
| `.count(sub)` | Number of occurrences | `0` |
| `.startswith(s)` | `bool` | — |
| `.endswith(s)` | `bool` | — |
| `sub in s` | `bool` | — |

````python
bio = "I love Python and Python loves me."

print(bio.count("Python"))      # 2
print(bio.find("Python"))       # 7
print(bio.find("Java"))         # -1  (safe — no error)

print(bio.startswith("I love")) # True
print(bio.endswith("me."))      # True
print(bio.startswith("You"))    # False

print("Python" in bio)          # True
print("Java" in bio)            # False
````

## Section 7 — f-String Format Specifications

Inside `{}` you can add a **format spec** after a colon: `{value:spec}`

| Spec | Meaning | Example |
|---|---|---|
| `:<10` | Left-align, width 10 | `'Berke     '` |
| `:>10` | Right-align, width 10 | `'     Berke'` |
| `:^10` | Center, width 10 | `'  Berke   '` |
| `:*^10` | Center, fill `*` | `'**Berke***'` |
| `:.2f` | 2 decimal places | `'95.68'` |
| `:,` | Thousands separator | `'1,000,000'` |

````python
name = "Berke"
score = 95.678
pi = 3.14159265

print(f"|{name:<10}|")          # |Berke     |
print(f"|{name:>10}|")          # |     Berke|
print(f"|{name:^10}|")          # |  Berke   |
print(f"|{name:*^10}|")         # |**Berke***|

print(f"{score:.2f}")           # 95.68
print(f"{pi:.4f}")              # 3.1416

big_number = 1_000_000
print(f"{big_number:,}")        # 1,000,000
print(f"{big_number:_}")        # 1_000_000
````

## Section 8 — len() and String Immutability

- `len(s)` → number of characters (spaces counted)
- Strings are **immutable**: you cannot assign to an index
- To "change" part of a string, build a new one with slicing and concatenation

````python
greeting = "Hello"
print(len(greeting))            # 5

long_str = "Hello, World!"
print(len(long_str))            # 13

# greeting[0] = "J"             # TypeError — uncomment to see the error

new_greeting = "J" + greeting[1:]
print(new_greeting)             # Jello
print(greeting)                 # Hello — original unchanged
````
