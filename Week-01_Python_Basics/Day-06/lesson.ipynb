## Section 1 — print() Parameters

```python
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```

| Parameter | Default | Purpose |
|-----------|---------|--------|
| `sep` | `' '` | String placed **between** each object |
| `end` | `'\n'` | String placed **after** the last object |
| `file` | `sys.stdout` | Where to write output |
| `flush` | `False` | Force-flush the output buffer |


````python
print("apple", "banana", "cherry")              # apple banana cherry
print("apple", "banana", "cherry", sep=", ")    # apple, banana, cherry
print("apple", "banana", "cherry", sep=" | ")   # apple | banana | cherry
print("apple", "banana", "cherry", sep="")      # applebananacherry

# end — stay on the same line
print("Loading", end="")
print("...", end="")
print(" Done!")             # Loading... Done!

# Combining sep and end
print("X", "Y", "Z", sep="-", end=" OK\n")   # X-Y-Z OK

# Blank line
print()                     # empty line
````

## Section 2 — f-strings Review

f-strings embed any Python expression directly in a string.

```python
f"text {expression} text"
```

Any valid Python expression works inside `{}`: variables, arithmetic, function calls, list indexing, etc.

````python
name  = "Berke"
score = 95
pi    = 3.14159

print(f"Hello, {name}!")             # Hello, Berke!
print(f"Score: {score}")             # Score: 95
print(f"2 + 2 = {2 + 2}")           # 2 + 2 = 4
print(f"Pi rounded: {round(pi, 2)}") # Pi rounded: 3.14

items = ["pen", "book", "ruler"]
print(f"First item: {items[0]}")     # First item: pen
print(f"Item count: {len(items)}")   # Item count: 3
````

## Section 3 — f-string Format Specs

```
{value : [fill][align][sign][#][0][width][grouping][.precision][type]}
```

| Part | Options | Effect |
|------|---------|--------|
| `align` | `<` `>` `^` | left / right / center |
| `fill` | any char | padding character (used with align) |
| `width` | integer | minimum field width |
| `.precision` | integer | decimal places (float) or max chars (str) |
| `type` | `d f e % b x o s` | int / float / sci / percent / binary / hex / octal / str |
| `#` | flag | add `0b` / `0o` / `0x` prefix |
| `,` | grouping | thousands separator |

````python
pi = 3.14159

# Alignment
print(f"{'left':<10}|")         # left      |
print(f"{'right':>10}|")        #      right|
print(f"{'center':^10}|")       #   center  |
print(f"{'fill':*^10}|")        # ***fill***|

# Width and precision
print(f"{pi:10.3f}")            #      3.142
print(f"{pi:.2f}")              # 3.14
print(f"{'hello':.3s}")         # hel  (max 3 chars)

# Sign
print(f"{42:+d}")               # +42
print(f"{-42:+d}")              # -42
print(f"{42: d}")               #  42  (space for positive)

# Zero padding
print(f"{42:06d}")              # 000042
print(f"{3.14:08.2f}")         # 00003.14

# Number types
n = 255
print(f"{n:d}")                 # 255
print(f"{n:b}")                 # 11111111
print(f"{n:o}")                 # 377
print(f"{n:x}")                 # ff
print(f"{n:X}")                 # FF
print(f"{n:#b}")                # 0b11111111
print(f"{n:#x}")                # 0xff
````

## Section 4 — format() Method

Uses the same format spec syntax as f-strings. Useful for building reusable template strings.

```python
"template {}".format(value)
"template {0} {1} {0}".format(a, b)       # positional
"template {name}".format(name=value)       # keyword
```

````python
# Positional
print("{} + {} = {}".format(1, 2, 3))          # 1 + 2 = 3
print("{0} and {1} and {0}".format("A", "B"))  # A and B and A

# Keyword
print("{name} scored {pts} points".format(name="Berke", pts=95))

# Format spec inside format()
print("{:.2f}".format(3.14159))     # 3.14
print("{:>10}".format("hi"))        #         hi
print("{:,}".format(1234567))       # 1,234,567

# Reusable template
template = "Product: {item:<15} Price: ${price:>8.2f}"
print(template.format(item="Laptop", price=999.9))
print(template.format(item="Mouse",  price=29.99))
````

## Section 5 — % Formatting (Legacy)

Older style still found in many codebases. Uses `%` as a placeholder marker.

| Code | Type |
|------|------|
| `%s` | string |
| `%d` | integer |
| `%f` | float |
| `%e` | scientific |
| `%x` | hex |

> Use f-strings in new code — they are cleaner, faster, and more powerful.

````python
name = "Berke"
age  = 25
gpa  = 3.87

print("Name: %s" % name)                # Name: Berke
print("Age: %d" % age)                  # Age: 25
print("GPA: %.2f" % gpa)               # GPA: 3.87
print("%s is %d years old." % (name, age))  # Berke is 25 years old.

# Width and alignment with %
print("%10s" % "right")                #      right
print("%-10s|" % "left")               # left      |
print("%08.2f" % 3.14)                 # 00003.14
````

## Section 6 — Number Formatting

| Format spec | Output | Description |
|-------------|--------|-------------|
| `{n:,}` | `1,234,567` | Thousands comma separator |
| `{n:_}` | `1_234_567` | Underscore separator |
| `{n:.2%}` | `87.35%` | Percentage (×100 + %) |
| `{n:.2e}` | `1.23e-06` | Scientific notation |
| `{n:#b}` | `0b1010` | Binary with prefix |
| `{n:#x}` | `0xff` | Hex with prefix |
| `${n:,.2f}` | `$1,234.50` | Currency style |

````python
big = 1_234_567_890
print(f"{big:,}")           # 1,234,567,890
print(f"{big:_}")           # 1_234_567_890

ratio = 0.8735
print(f"{ratio:.1%}")       # 87.4%
print(f"{ratio:.2%}")       # 87.35%

tiny  = 0.000001234
large = 9_870_000.0
print(f"{tiny:.2e}")        # 1.23e-06
print(f"{large:.3E}")       # 9.870E+06

price = 1234.5
print(f"${price:,.2f}")     # $1,234.50

byte = 0b10110011
print(f"decimal : {byte:d}")
print(f"binary  : {byte:08b}")
print(f"hex     : {byte:02x}")
print(f"octal   : {byte:o}")
````

## Section 7 — String Alignment

| Method | Action |
|--------|--------|
| `s.ljust(w, c)` | Pad right with `c` to reach width `w` |
| `s.rjust(w, c)` | Pad left with `c` |
| `s.center(w, c)` | Center with `c` on both sides |
| `s.zfill(w)` | Pad with zeros on the left (sign stays at front) |

f-string alignment specs are preferred in modern code — they are more flexible and composable.

````python
word = "Python"
print(word.ljust(12))            # Python        
print(word.ljust(12, "-"))       # Python------
print(word.rjust(12))            #       Python
print(word.rjust(12, "."))       # ......Python
print(word.center(12))           #    Python   
print(word.center(12, "="))      # ===Python===
print("42".zfill(6))             # 000042
print("-7".zfill(6))             # -00007

# Aligned table with f-strings
print()
print(f"{'Name':<12} {'Score':>6} {'Grade':^8}")
print("-" * 28)
data = [("Alice", 95, "A"), ("Bob", 72, "C"), ("Charlie", 88, "B")]
for name, sc, grade in data:
    print(f"{name:<12} {sc:>6} {grade:^8}")
````

## Section 8 — Print Tricks

| Technique | How |
|-----------|-----|
| Stay on same line | `end=""` |
| Rewrite current line | `end="\r"` (carriage return) |
| Force output immediately | `flush=True` |
| Dynamic width in f-string | `f"{'text':^{width}}"` |

`\r` moves the cursor back to the start of the **current** line without advancing — perfect for progress bars.

````python
import time

# Rewriting the same line — progress bar simulation
BAR = 30
for step in range(11):
    pct    = step * 10
    filled = int(BAR * pct / 100)
    bar    = "\u2588" * filled + "\u2591" * (BAR - filled)
    print(f"\r[{bar}] {pct:3d}%", end="", flush=True)
    time.sleep(0.1)
print("\nDone!")

# Dynamic width using a variable in the format spec
WIDTH = 50
print("=" * WIDTH)
print(f"{'REPORT':^{WIDTH}}")
print("=" * WIDTH)

# Decorative box function
def print_box(text):
    border = "+" + "-" * (len(text) + 2) + "+"
    print(border)
    print(f"| {text} |")
    print(border)

print_box("Hello, Python!")
print_box("Week 1 — Day 6")
````

## Quick Reference Cheatsheet

```python
# print() parameters
print(a, b, sep="-")     # a-b
print(a, end="")         # no newline after a
print(a, end="\r")       # rewrite current line

# f-string format spec: {value:[fill][align][sign][0][width][,][.prec][type]}
f"{n:,}"          # 1,234,567  (thousands comma)
f"{n:.2f}"        # 3.14       (2 decimal places)
f"{n:.2%}"        # 87.35%     (percentage)
f"{n:.2e}"        # 1.23e+02   (scientific)
f"{n:08d}"        # 00000042   (zero-padded)
f"{n:#b}"         # 0b101010   (binary with prefix)
f"{n:#x}"         # 0xff       (hex with prefix)
f"{s:<10}"        # left-align in 10 chars
f"{s:>10}"        # right-align
f"{s:^10}"        # center
f"{s:*^10}"       # center with * fill
f"{s:.3s}"        # max 3 chars

# Dynamic width in f-string
f"{'text':^{width}}"    # width is a variable

# format() — same spec, reusable templates
"{:>10}".format("hi")         #         hi
"{name}".format(name="Berke") # Berke

# str alignment methods
s.ljust(w, c)    s.rjust(w, c)    s.center(w, c)    s.zfill(w)
```

> ⚠️ **Common traps:**
> - `{n:.2%}` multiplies by 100 first — `0.5` → `50.00%`, not `0.50%`
> - `end="\r"` without `flush=True` may not display in all environments
> - `%` formatting requires a tuple for multiple values: `"%s %d" % (a, b)` not `% a, b`
