## Section 1 — Numeric Types

Python has three built-in numeric types:

| Type | Description | Example |
|------|-------------|--------|
| `int` | Whole numbers, no decimal, unlimited size | `42`, `-7`, `1_000_000` |
| `float` | Numbers with a decimal point (IEEE 754) | `3.14`, `2e8`, `9.11e-31` |
| `complex` | Numbers with a real and imaginary part | `3+2j` |

> **Dynamic typing:** Python determines the type automatically — you never declare it manually.

````python
age      = 25               # int
price    = 19.99            # float
ratio    = 1 / 3            # float — any division produces a float
big_int  = 1_000_000        # int — underscores are visual separators only

print(type(age))            # <class 'int'>
print(type(price))          # <class 'float'>
print(type(ratio))          # <class 'float'>
print(type(1 + 2j))         # <class 'complex'>

# Scientific notation is valid float syntax
speed_of_light = 3e8        # 300000000.0
electron_mass  = 9.11e-31
print(speed_of_light)       # 300000000.0
print(electron_mass)        # 9.11e-31
````

## Section 2 — Arithmetic Operators

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `3 + 2` | `5` |
| `-` | Subtraction | `3 - 2` | `1` |
| `*` | Multiplication | `3 * 2` | `6` |
| `/` | True division | `7 / 2` | `3.5` (always float) |
| `//` | Floor division | `7 // 2` | `3` |
| `%` | Modulo | `7 % 2` | `1` |
| `**` | Exponentiation | `2 ** 3` | `8` |

````python
a = 17
b = 5

print(a + b)                # 22
print(a - b)                # 12
print(a * b)                # 85
print(a / b)                # 3.4   (always float)
print(a // b)               # 3     (floor division)
print(a % b)                # 2     (remainder)
print(a ** b)               # 1419857

# True division ALWAYS returns float
print(10 / 2)               # 5.0   (not 5)
print(10 / 3)               # 3.3333333333333335
````

---
## Section 3 — Operator Precedence

When multiple operators appear in one expression, Python follows this order (highest to lowest):

| Priority | Operator | Name |
|----------|----------|------|
| 1 (highest) | `()` | Parentheses |
| 2 | `**` | Exponentiation (right-to-left) |
| 3 | `+x`, `-x` | Unary plus / minus |
| 4 | `*`, `/`, `//`, `%` | Multiply / Divide / Floor / Mod |
| 5 (lowest) | `+`, `-` | Addition / Subtraction |

> **Rule of thumb:** When in doubt, add parentheses — they never hurt clarity.

````python
print(2 + 3 * 4)            # 14   (* before +)
print((2 + 3) * 4)          # 20   (parens override)
print(2 ** 3 ** 2)          # 512  (right-to-left: 3**2=9, then 2**9=512)
print(10 - 4 + 2)           # 8    (left-to-right)
print(10 - (4 + 2))         # 4    (parens change the result)
````

## Section 4 — Integer Division & Modulo

- `//` (floor division) rounds toward **negative infinity** — not toward zero.
- `%` (modulo) gives the remainder consistent with floor division.
- The result of `a % b` always has the **same sign as `b`** (the divisor).

**Practical uses of modulo:**
- Check even/odd: `n % 2 == 0`
- Wrap around a range (clock arithmetic): `hour % 12`
- Extract a digit: `number % 10` gives the last digit

````python
# Floor division — rounds toward negative infinity, not toward zero
print(7 // 2)               # 3
print(-7 // 2)              # -4   (NOT -3)
print(7 // -2)              # -4

# Modulo — result sign always matches the divisor
print(17 % 5)               # 2    (17 = 3*5 + 2)
print(-17 % 5)              # 3    (not -2)
print(17 % -5)              # -3

# Practical: even/odd check
print(10 % 2 == 0)          # True  (even)
print(7  % 2 == 0)          # False (odd)

# Practical: clock arithmetic
hour = 14
print(hour % 12)            # 2     (2 PM on a 12-hour clock)

# Practical: extract digits
number = 12345
print(number % 10)          # 5     (last digit)
print(number // 10 % 10)    # 4     (second-to-last digit)
````

## Section 5 — Built-in Math Functions

These functions are always available — no import needed:

| Function | What it does |
|----------|--------------|
| `abs(x)` | Absolute value |
| `round(x, n)` | Round to `n` decimal places |
| `pow(x, y)` | `x**y` — or `pow(x, y, mod)` for modular form |
| `divmod(a, b)` | Returns `(a//b, a%b)` as a tuple |
| `min(...)` | Smallest value |
| `max(...)` | Largest value |
| `sum(...)` | Sum of an iterable |

````python
print(abs(-42))             # 42
print(abs(3.7))             # 3.7

print(round(3.14159, 2))    # 3.14
print(round(3.14159, 4))    # 3.1416
print(round(2.5))           # 2   (banker's rounding — rounds to even)
print(round(3.5))           # 4

print(pow(2, 10))           # 1024
print(pow(2, 10, 1000))     # 24   ((2**10) % 1000)

print(min(3, 1, 4, 1, 5))   # 1
print(max(3, 1, 4, 1, 5))   # 5
print(sum([1, 2, 3, 4, 5])) # 15

# divmod returns (quotient, remainder) in one call
q, r = divmod(17, 5)
print(q, r)                 # 3 2
````

## Section 6 — The `math` Module

The standard library `math` module provides constants and functions for advanced calculations.

**Constants:** `math.pi`, `math.e`, `math.inf`, `math.tau`  
**Rounding:** `math.floor()`, `math.ceil()`, `math.trunc()`  
**Power/log:** `math.sqrt()`, `math.log()`, `math.log10()`, `math.log2()`  
**Trig:** `math.sin()`, `math.cos()`, `math.degrees()`, `math.radians()`  
**Other:** `math.factorial()`, `math.gcd()`, `math.isfinite()`, `math.isnan()`

````python
import math

# Constants
print(math.pi)              # 3.141592653589793
print(math.e)               # 2.718281828459045
print(math.inf)             # inf
print(math.tau)             # 6.283185307179586  (2 * pi)

# Rounding variants
print(math.floor(3.9))      # 3    (toward -infinity)
print(math.ceil(3.1))       # 4    (toward +infinity)
print(math.trunc(3.9))      # 3    (toward zero)
print(math.trunc(-3.9))     # -3   (toward zero, not -4)

# Power and logarithms
print(math.sqrt(144))       # 12.0
print(math.log(math.e))     # 1.0
print(math.log(100, 10))    # 2.0
print(math.log10(1000))     # 3.0
print(math.log2(256))       # 8.0

# Trigonometry (angles in radians)
print(math.sin(math.pi / 2))  # 1.0
print(math.cos(0))            # 1.0
print(math.degrees(math.pi))  # 180.0
print(math.radians(90))       # 1.5707963267948966

# Combinatorics & number theory
print(math.factorial(5))    # 120
print(math.gcd(48, 18))     # 6

print(math.isfinite(math.inf))   # False
print(math.isnan(float("nan")))  # True
````

## Section 7 — Type Conversion

| Function | Converts to | Notes |
|----------|-------------|-------|
| `int(x)` | integer | Truncates floats (does NOT round); parses numeric strings |
| `float(x)` | float | Promotes int; parses numeric strings |
| `int(s, base)` | integer | Parse string in given base (2 = binary, 16 = hex, 8 = octal) |
| `bin(n)` | string | Decimal → binary string (`0b...`) |
| `hex(n)` | string | Decimal → hex string (`0x...`) |
| `oct(n)` | string | Decimal → octal string (`0o...`) |

> ⚠️ **Float precision:** `0.1 + 0.2` does not equal `0.3` exactly — this is an IEEE 754 property, not a Python bug. Use `round()` when exact comparison matters.

````python
print(int(3.9))             # 3    (truncates — does NOT round)
print(int(-3.9))            # -3   (truncates toward zero)
print(float(7))             # 7.0

# Strings that look like numbers can be converted
print(int("42"))            # 42
print(float("3.14"))        # 3.14

# int() accepts a base argument for number-system conversions
print(int("1010", 2))       # 10   (binary  -> decimal)
print(int("FF", 16))        # 255  (hex     -> decimal)
print(int("17", 8))         # 15   (octal   -> decimal)

# Going the other direction:
print(bin(10))              # 0b1010
print(hex(255))             # 0xff
print(oct(15))              # 0o17

# Float arithmetic has limited precision (IEEE 754)
print(0.1 + 0.2)            # 0.30000000000000004
print(round(0.1 + 0.2, 2))  # 0.3
````

## Section 8 — Augmented Assignment Operators

Shorthand that reads **"update this variable using itself"**:

| Shorthand | Equivalent to |
|-----------|---------------|
| `x += n` | `x = x + n` |
| `x -= n` | `x = x - n` |
| `x *= n` | `x = x * n` |
| `x /= n` | `x = x / n` |
| `x //= n` | `x = x // n` |
| `x %= n` | `x = x % n` |
| `x **= n` | `x = x ** n` |

````python
score = 100
score += 10                 # score = score + 10
print(score)                # 110

score -= 5
print(score)                # 105

score *= 2
print(score)                # 210

score //= 3
print(score)                # 70

score **= 2
print(score)                # 4900

score %= 100
print(score)                # 0
````
