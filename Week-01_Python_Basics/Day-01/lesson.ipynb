## Section 1 — What Is a Variable?

A **variable** is a named container that holds a value in memory.  
Think of it like a labelled box: you put something inside, then refer to the box by its label.

Python has **5 core data types** you'll use every day:

| Type | Example | Used For |
|------|---------|----------|
| `int` | `25` | Whole numbers |
| `float` | `1.82` | Decimal numbers |
| `str` | `"Berke"` | Text |
| `bool` | `True` | Yes/No logic |
| `NoneType` | `None` | No value / missing |

````python
name = "Berke"          # str
age = 25                # int
height = 1.82           # float
is_developer = True     # bool
middle_name = None      # NoneType

print(name)
print(age)
print(height)
print(is_developer)
print(middle_name)
````

## Section 2 — The `type()` Function

Python is **dynamically typed** — it figures out the type automatically.  
Use `type()` to inspect what Python decided.

````python
print(type(name))           # <class 'str'>
print(type(age))            # <class 'int'>
print(type(height))         # <class 'float'>
print(type(is_developer))   # <class 'bool'>
print(type(middle_name))    # <class 'NoneType'>
````

## Section 3 — Naming Rules & PEP 8 Conventions

**Rules (enforced by Python):**
- Letters, digits, underscores only
- Cannot start with a digit
- Cannot use keywords (`if`, `for`, `class`, …)
- Case-sensitive: `name ≠ Name ≠ NAME`

**Conventions (PEP 8):**
- Variables/functions → `snake_case`
- Constants → `UPPER_SNAKE_CASE`
- Classes → `PascalCase`

````python
first_name = "Berke"        # good
last_name = "Turk"          # good
MAX_RETRIES = 3             # good — constant

print(first_name, last_name)
print(MAX_RETRIES)
````

## Section 4 — Reassignment & Dynamic Typing

Variables can be reassigned — even to a **different type**.  
Python updates the type automatically.

````python
counter = 10
print(counter, type(counter))   # 10 <class 'int'>

counter = "ten"
print(counter, type(counter))   # ten <class 'str'>
````

## Section 5 — Multiple Assignment

````python
# Same value to multiple variables
x = y = z = 0
print(x, y, z)                  # 0 0 0

# Tuple unpacking — assign different values in one line
first, second, third = "Python", 3.11, True
print(first, second, third)     # Python 3.11 True
````

## Section 6 — The `id()` Function

`id()` returns the unique **memory address** of an object.  
Python caches small integers (-5 to 256), so they share addresses.

````python
a = 5
b = 5
print(id(a) == id(b))           # True — Python reuses small int objects

big_a = 1000
big_b = 1000
print(id(big_a) == id(big_b))   # False — large ints get new objects
````

## Section 7 — String Basics

````python
greeting = "Hello, World!"
language = 'Python'             # single or double quotes — identical
bio = """I am Berke.
I am a developer.
I love Python."""

print(greeting)
print(language)
print(bio)
````

## Section 8 — f-Strings

**f-strings** are the modern way to embed variables inside strings.  
Prefix with `f` and wrap expressions in `{}`.

````python
name = "Berke"
age = 25
language = "Python"

intro = f"Hi, I'm {name}. I'm {age} years old and I code in {language}."
print(intro)

# Expressions inside {} are evaluated
print(f"In 10 years I will be {age + 10} years old.")
````
