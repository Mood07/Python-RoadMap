## Section 1 — The `input()` Function

`input("prompt")` displays a prompt, waits for the user to press Enter, and returns what they typed as a **string — always `str`**.

```python
# Real usage pattern:
name = input("Enter your name: ")
# User types: Berke
# name is now the string "Berke"
```

> **Key rule:** Even if the user types `42`, `input()` returns the string `"42"`. You must convert it yourself.

````python
# Simulating: name = input("Enter your name: ")
name = "Berke"
print(f"Hello, {name}!")
print(type(name))               # <class 'str'>

# Even digits come back as str
raw_number = "25"               # simulates: input("Type any number: ")
print(type(raw_number))         # <class 'str'>
print(raw_number + "10")        # 2510  — string concat, NOT math!
````

---
## Section 2 — Converting Input to `int`

Wrap `input()` with `int()` to get a whole number ready for arithmetic.

```python
age = int(input("Enter your age: "))
```

| Expression | Result | Why |
|------------|--------|-----|
| `"25" + "10"` | `"2510"` | String concatenation |
| `int("25") + 10` | `35` | Integer addition |
| `int("3.14")` | `ValueError` | int() rejects decimal strings |
| `int("  42  ")` | `42` | int() strips whitespace automatically |

````python
age_str = "25"                  # simulates: input("Enter your age: ")
age = int(age_str)

print(f"You are {age} years old.")
print(f"In 10 years: {age + 10}")
print(f"Type: {type(age)}")

# Demonstrating the difference
print("25" + "10")              # 2510 — string concat
print(int("25") + 10)          # 35   — integer math
print(int("  42  "))            # 42   — whitespace stripped
````

---
## Section 3 — Converting Input to `float`

Use `float()` when the user might enter a decimal number.

```python
price = float(input("Enter price: "))
```

| Expression | Result |
|------------|--------|
| `float("3")` | `3.0` — accepts integer strings |
| `float("3.14")` | `3.14` — accepts decimal strings |
| `int("3.14")` | `ValueError` — int() rejects decimal strings |

````python
price_str = "19.99"             # simulates: input("Enter price: ")
qty_str   = "3"                 # simulates: input("Enter quantity: ")

price    = float(price_str)
quantity = int(qty_str)
total    = price * quantity

print(f"Price   : {price}")
print(f"Qty     : {quantity}")
print(f"Total   : {total:.2f}")

# float() accepts both int and decimal strings
print(float("3"))               # 3.0
print(float("3.14"))            # 3.14
````

## Section 4 — `str()` — Explicit String Conversion

`str()` converts **any** Python value to its string representation. Useful when concatenating non-strings with `+`.

| Expression | Result |
|------------|--------|
| `"Score: " + 95` | `TypeError` — can't concat str + int |
| `"Score: " + str(95)` | `"Score: 95"` |
| `str(3.14)` | `"3.14"` |
| `str(True)` | `"True"` |
| `str(None)` | `"None"` |

````python
score = 95
grade = "A"

# Using str() with concatenation
message = "Score: " + str(score) + " | Grade: " + grade
print(message)

# f-strings are cleaner (they call str() automatically)
print(f"Score: {score} | Grade: {grade}")

# str() on other types
print(str(3.14))                # 3.14
print(str(True))                # True
print(str(None))                # None
print(str([1, 2, 3]))           # [1, 2, 3]
````

## Section 5 — `bool()` and Truthiness

`bool()` converts a value to `True` or `False`. Python defines **falsy** values that behave like `False` in conditions — everything else is **truthy**.

| Falsy values | Truthy examples |
|---|---|
| `0`, `0.0`, `0j` | Any non-zero number |
| `""` (empty string) | `"hello"`, `"0"`, `"False"` |
| `[]`, `{}`, `set()`, `()` | Any non-empty collection |
| `None`, `False` | Anything else |

> ⚠️ **Common trap:** `bool("False")` is `True` — because `"False"` is a **non-empty string**.

````python
# Falsy
print(bool(0))                  # False
print(bool(""))                 # False
print(bool(None))               # False
print(bool([]))                 # False

# Truthy
print(bool(1))                  # True
print(bool(-99))                # True  (non-zero)
print(bool("hello"))            # True
print(bool("False"))            # True  (non-empty string!)
print(bool("0"))                # True  (non-empty string!)

# Correct way to check yes/no from user input
response = "yes"                # simulates: input("Continue? ")
if response.lower() in ("yes", "y"):
    print("Continuing...")
````

## Section 6 — Validation Patterns

User input is **untrusted**. Validate before converting to avoid crashes.

| Pattern | Use case |
|---------|----------|
| `.isdigit()` | Check if string is a non-negative integer |
| `.strip()` | Remove accidental whitespace before converting |
| `try / except ValueError` | Safe conversion for any numeric type |

````python
# isdigit() — only True for non-negative integer strings
print("42".isdigit())           # True
print("3.14".isdigit())         # False  (dot is not a digit)
print("-5".isdigit())           # False  (minus is not a digit)
print("".isdigit())             # False  (empty)

# strip() before converting
messy = "  99  "
print(int(messy.strip()))       # 99

# try/except — safe conversion helper
def safe_int(s):
    try:
        return int(s)
    except ValueError:
        return None

print(safe_int("42"))           # 42
print(safe_int("hello"))        # None
````

## Section 7 — Multiple Inputs

Collect several values in one line using `split()`.

```python
# Split on whitespace
a, b = input("Enter two numbers: ").split()

# Split on a specific delimiter
first, last = input("First, Last: ").split(",")

# Convert all at once with map()
nums = list(map(int, input("Numbers: ").split()))
```

````python
# Split on whitespace — basic
line = "10 20"                  # simulates: input("Enter two numbers: ")
parts = line.split()
x, y = int(parts[0]), int(parts[1])
print(f"{x} + {y} = {x + y}")  # 10 + 20 = 30

# Pythonic one-liner with unpacking
raw = "15 25"                   # simulates: input("Enter two numbers: ")
a, b = raw.split()
a, b = int(a), int(b)
print(f"{a} * {b} = {a * b}")  # 15 * 25 = 375

# Split on a comma delimiter
raw = "Berke, 25"               # simulates: input("Name, Age: ")
name_part, age_part = raw.split(",")
print(f"Name: {name_part.strip()} | Age: {int(age_part.strip())}")

# map() converts all parts at once
raw = "3 7 11"
nums = list(map(int, raw.split()))
print(nums)                     # [3, 7, 11]
print(sum(nums))                # 21
````

## Section 8 — Complete Program Pattern

A well-structured input program follows this order:
1. **Collect** all input at the top
2. **Convert** and validate
3. **Compute** derived values
4. **Display** results

````python
# Simulated profile builder (replace hardcoded strings with input() to run live)
full_name  = "Berke Arda Turk"  # input("Full name  : ")
birth_year = "2000"             # input("Birth year : ")
height_cm  = "182"              # input("Height (cm): ")
city       = "Istanbul"         # input("City       : ")

# Convert
birth_year_int = int(birth_year)
height_m       = int(height_cm) / 100

# Compute
age          = 2025 - birth_year_int
retire_years = 65 - age

# Display
print("=" * 38)
print("  PROFILE CARD")
print("=" * 38)
print(f"  Name       : {full_name}")
print(f"  Age        : {age}")
print(f"  Height     : {height_m:.2f} m")
print(f"  City       : {city}")
print(f"  Born       : {birth_year_int}")
print(f"  Retirement : in {retire_years} years")
print("=" * 38)
````

## Quick Reference Cheatsheet

```python
# input() always returns str
name = input("prompt")          # str
age  = int(input("prompt"))     # str -> int
price = float(input("prompt"))  # str -> float

# Conversion functions
int(x)    # truncates float, parses str, ValueError on bad input
float(x)  # promotes int, parses str, ValueError on bad input
str(x)    # works on any type, never raises
bool(x)   # False for: 0  0.0  ""  []  {}  ()  None  False

# Validation
s.isdigit()           # True only for non-negative integer strings
s.strip()             # remove leading/trailing whitespace
try:
    n = int(s)
except ValueError:
    ...               # handle bad input safely

# Multiple inputs
a, b = input("...").split()           # split on whitespace
a, b = input("...").split(",")        # split on comma
nums = list(map(int, s.split()))      # split + convert all
```

> ⚠️ **Common traps:**
> - `bool("False")` is `True` — non-empty string is always truthy
> - `int("3.14")` raises `ValueError` — use `float()` first, then `int()`
> - `"25" + 10` raises `TypeError` — convert with `int()` before arithmetic
