## Part 1 â€” Multiple Choice Questions

### Day 1 â€” Variables & Data Types

**Q1:** What does `type(3.0)` return?

- A) `<class 'int'>`
- B) `<class 'float'>` âś…
- C) `<class 'str'>`
- D) `<class 'number'>`

> `3.0` has a decimal point, so Python stores it as `float`. `type()` returns the class of the object.

---

**Q2:** What does `bool([])` return?

- A) `True`
- B) `None`
- C) `False` âś…
- D) `0`

> An empty list `[]` is **falsy**. `bool()` on any empty collection (`[]`, `{}`, `set()`, `()`) returns `False`.

### Day 2 â€” String Operations

**Q3:** What does `"hello"[1:4]` return?

- A) `"hell"`
- B) `"ell"` âś…
- C) `"hel"`
- D) `"ello"`

> Slicing `[1:4]` starts at index 1 (`e`) and stops **before** index 4. Indices 1,2,3 â†’ `'e','l','l'` â†’ `'ell'`.

---

**Q4:** What does `"  hello  ".strip()` return?

- A) `"  hello  "` (unchanged)
- B) `"hello"` âś…
- C) `"hello  "`
- D) `"  hello"`

> `strip()` removes **all** leading and trailing whitespace from both ends. `lstrip()` is left only, `rstrip()` is right only.

### Day 3 â€” Numbers & Math

**Q5:** What does `17 % 5` evaluate to?

- A) `3`
- B) `2` âś…
- C) `3.4`
- D) `1`

> `%` is the **modulo** (remainder) operator. 17 Ă· 5 = 3 remainder **2**. Useful for even/odd checks, cycling, etc.

---

**Q6:** What does `int(3.9)` return?

- A) `4` (rounds up)
- B) `3` (truncates) âś…
- C) `3.9`
- D) `ValueError`

> `int()` **truncates** â€” it cuts off the decimal part without rounding. `int(3.9)` â†’ `3`, `int(-3.9)` â†’ `-3`. Use `round()` to round.

### Day 4 â€” User Input & Type Conversion

**Q7:** What type does `input()` **always** return?

- A) `int`
- B) `float`
- C) `str` âś…
- D) depends on what the user types

> `input()` **always** returns `str`, no matter what the user types. Even `42` comes back as `"42"`. You must convert explicitly.

---

**Q8:** What does `"3.14".isdigit()` return?

- A) `True`
- B) `False` âś…
- C) `ValueError`
- D) `None`

> `isdigit()` returns `True` only for strings containing **only** digit characters (0â€“9). The dot `'.'` is not a digit â†’ `False`. Negative signs also cause `False`.

### Day 5 â€” Operators & Expressions

**Q9:** What does `5 | 3` (bitwise OR) evaluate to?

- A) `1`
- B) `7` âś…
- C) `2`
- D) `15`

> In binary: `5 = 101`, `3 = 011`. Bitwise OR gives 1 where **at least one** bit is 1: `101 | 011 = 111 = 7`.

---

**Q10:** What does `not (True and False)` evaluate to?

- A) `False`
- B) `True` âś…
- C) `None`
- D) `TypeError`

> Evaluate inside-out: `True and False` â†’ `False`. Then `not False` â†’ `True`. Parentheses force `and` to evaluate first.

### C2 â€” Day 3: Integer Division & Modulo

```python
a = 10
b = 3
print(a // b, a % b)
# Predict before running!
# Answer: 3 1
# 10 // 3 = 3 (floor division), 10 % 3 = 1 (remainder)
```

### C3 â€” Day 5: Ternary Expression

```python
x = 7
print("odd" if x % 2 else "even")
# Predict before running!
# Answer: odd
# 7 % 2 = 1, which is truthy â†’ ternary returns 'odd'
```

### C4 â€” Day 6: f-string Center Fill

```python
print(f"{'hi':*^8}")
# Predict before running!
# Answer: ***hi***
# *^8 â†’ center in width 8, fill with *. 'hi' is 2 chars â†’ 3 stars each side.
```

### C5 â€” Day 6: print() sep & end

```python
print("A", "B", sep="-", end="!")
print("C")
# Predict before running!
# Answer: A-B!C
# First print: sep='-' â†’ 'A-B', end='!' â†’ no newline
# Second print: 'C' then default newline â†’ cursor moves down
# Combined on one line: A-B!C
```
