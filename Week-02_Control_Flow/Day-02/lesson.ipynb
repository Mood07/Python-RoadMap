# Week 2 - Day 2: for Loops
**Topic:** range(), iterate lists/strings/dicts, enumerate(), zip()

## 1. Basic for Loop
Runs the block once for each item in the iterable.

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

## 2. range() Function
| Form | Produces |
|---|---|
| `range(stop)` | 0, 1, …, stop-1 |
| `range(start, stop)` | start, …, stop-1 |
| `range(start, stop, step)` | every step-th value |

```python
for i in range(5):
    print(i, end=" ")       # 0 1 2 3 4
print()

for i in range(1, 6):
    print(i, end=" ")       # 1 2 3 4 5
print()

for i in range(0, 11, 2):
    print(i, end=" ")       # 0 2 4 6 8 10
print()

for i in range(5, 0, -1):
    print(i, end=" ")       # 5 4 3 2 1
print()
```

## 3. Iterating Over a String
Strings are sequences — each character is one iteration.

```python
for char in "Python":
    print(char, end="-")    # P-y-t-h-o-n-
print()

vowels = 0
for char in "Hello World":
    if char.lower() in "aeiou":
        vowels += 1
print(f"Vowels: {vowels}")  # 3
```

## 4. Iterating Over a Dictionary
- `for key in d` → keys only
- `for v in d.values()` → values only
- `for k, v in d.items()` → key-value pairs

```python
student = {"name": "Berke", "age": 22, "grade": "A"}

for key in student:
    print(key)                      # name  age  grade

print()
for key, value in student.items():
    print(f"{key}: {value}")
```

## 5. enumerate() — Index + Value Together
`enumerate(iterable, start=0)` avoids manual counter variables.

```python
colors = ["red", "green", "blue"]

for index, color in enumerate(colors):
    print(f"{index}: {color}")
# 0: red  1: green  2: blue

print()
for i, color in enumerate(colors, start=1):
    print(f"{i}. {color}")
# 1. red  2. green  3. blue
```

## 6. zip() — Iterate Two Sequences in Parallel
`zip(a, b)` pairs items at the same position. Stops at the shortest sequence.

```python
names  = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]

for name, score in zip(names, scores):
    print(f"{name}: {score}")
```

## 7. Accumulator Pattern
Build a result by appending or summing inside the loop.

```python
squares = []
for n in range(1, 6):
    squares.append(n ** 2)
print(squares)              # [1, 4, 9, 16, 25]

total = 0
for num in [10, 20, 30, 40]:
    total += num
print(f"Total: {total}")    # 100
```

## 8. Loop with else Clause
The `else` block runs after normal completion — **not** when interrupted by `break`.

```python
target = 7
for n in [1, 3, 5, 9]:
    if n == target:
        print("Found!")
        break
else:
    print("Not found.")     # Runs — 7 is not in the list
```

## 9. Nested for Loops

```python
for row in range(1, 4):
    for col in range(1, 4):
        print(f"({row},{col})", end=" ")
    print()
```

## 10. Practical Example — Grade Report

```python
students = [
    {"name": "Alice",   "scores": [85, 90, 92]},
    {"name": "Bob",     "scores": [70, 65, 80]},
    {"name": "Charlie", "scores": [95, 98, 100]},
]

print("── Grade Report ──")
for i, student in enumerate(students, start=1):
    avg = sum(student["scores"]) / len(student["scores"])
    grade = "A" if avg >= 90 else "B" if avg >= 80 else "C"
    print(f"{i}. {student['name']:<10} Avg: {avg:.1f}  Grade: {grade}")
```
