# Week 2 - Day 4: break / continue / pass
**Topic:** Loop control statements

## 1. break — Exit the Loop Immediately
`break` stops the loop entirely. Execution jumps to the first line after the loop.

```python
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")       # 0 1 2 3 4
print()

count = 0
while True:
    if count == 3:
        break
    print(count, end=" ")   # 0 1 2
    count += 1
print()
```

## 2. continue — Skip the Current Iteration
`continue` skips the rest of the current iteration and jumps back to the condition check.

```python
for i in range(8):
    if i % 2 == 0:
        continue            # skip even numbers
    print(i, end=" ")       # 1 3 5 7
print()
```

## 3. pass — Do Nothing (Placeholder)
`pass` is a no-op. Use it when Python requires a statement but you have nothing to put there yet.

```python
for i in range(5):
    pass                    # loop runs, does nothing visible

def future_feature():
    pass                    # to be implemented later

class MyModel:
    pass                    # empty class skeleton
```

## 4. break vs continue — Side by Side

```python
numbers = [1, 8, 2, 9, 3]

print("break — stops at first >6:")
for n in numbers:
    if n > 6:
        break
    print(n, end=" ")       # 1  → stops at 8
print()

print("continue — skips all >6:")
for n in numbers:
    if n > 6:
        continue
    print(n, end=" ")       # 1 2 3
print()
```

## 5. break in Nested Loops
`break` only exits the **innermost** loop. Use a flag variable to break outer loops.

```python
for i in range(3):
    for j in range(3):
        if j == 1:
            break           # exits inner loop only
        print(f"({i},{j})", end=" ")
    print()

# Flag to break outer loop
found = False
for i in range(5):
    for j in range(5):
        if i * j == 6:
            found = True
            break
    if found:
        break
print(f"Found i={i}, j={j}")
```

## 6. for / else and while / else
`else` runs only if the loop was **not** exited via `break`. Useful for search patterns.

```python
def find_prime(numbers):
    for n in numbers:
        for divisor in range(2, n):
            if n % divisor == 0:
                break
        else:
            print(f"{n} is prime")

find_prime([7, 8, 11, 12, 13])
# 7 is prime  /  11 is prime  /  13 is prime
```

## 7. Practical Patterns

```python
# Search and stop
items = ["apple", "grape", "mango", "banana", "kiwi"]
for i, item in enumerate(items):
    if item == "mango":
        print(f"Found 'mango' at index {i}")
        break
else:
    print("Not found")

# Filter with continue
data = [15, -3, 42, -7, 0, 8, -1, 99]
print("Positive only:", end=" ")
for val in data:
    if val <= 0:
        continue
    print(val, end=" ")
print()
```

## 8. Practical Example — Text Processor

```python
def process_lines(lines):
    results = []
    for line in lines:
        line = line.strip()
        if not line:
            continue                # skip blank lines
        if line.startswith("#"):
            continue                # skip comments
        if line == "STOP":
            break                   # halt processing
        results.append(line.upper())
    return results

sample = [
    "# Header",
    "",
    "hello world",
    "  python is great  ",
    "loop control",
    "STOP",
    "never reached",
]

for line in process_lines(sample):
    print(line)
```
