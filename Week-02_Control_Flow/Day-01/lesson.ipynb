# Week 2 - Day 1: if / elif / else
**Topic:** Conditions, Nested if, Truthiness

## 1. Basic if Statement
Runs the indented block only when the condition is `True`.

````python
age = 20

if age >= 18:
    print("You are an adult.")   # Runs — 20 >= 18 is True

temperature = 5
if temperature < 0:
    print("It is freezing!")     # Does NOT run — condition is False
````

## 2. if / else
`else` runs when the `if` condition is `False`.

````python
score = 45

if score >= 50:
    print("Pass")
else:
    print("Fail")    # Runs — 45 < 50
````

## 3. if / elif / else
Python checks branches in order and stops at the **first True** one.

````python
grade = 78

if grade >= 90:
    print("A")
elif grade >= 80:
    print("B")
elif grade >= 70:
    print("C")    # Runs — 78 >= 70
elif grade >= 60:
    print("D")
else:
    print("F")
````

## 4. Comparison Operators
| Operator | Meaning |
|---|---|
| `==` | equal to |
| `!=` | not equal to |
| `>` | greater than |
| `<` | less than |
| `>=` | greater than or equal to |
| `<=` | less than or equal to |

````python
x = 10
print(x == 10)   # True
print(x != 5)    # True
print(x > 15)    # False
print(x <= 10)   # True
````

## 5. Logical Operators in Conditions
- `and` — both must be True
- `or`  — at least one must be True
- `not` — reverses the boolean

````python
age = 25
has_id = True

if age >= 18 and has_id:
    print("Entry allowed.")         # Runs

is_raining = False
has_umbrella = True
if is_raining or has_umbrella:
    print("You are prepared.")      # Runs

is_logged_in = False
if not is_logged_in:
    print("Please log in.")         # Runs
````

## 6. Truthiness — Truthy vs Falsy Values
**Falsy:** `False`, `None`, `0`, `0.0`, `""`, `[]`, `{}`, `set()`, `()`  
**Everything else** is truthy.

````python
name = ""
if name:
    print("Name is set.")
else:
    print("Name is empty.")    # Runs — empty string is falsy

items = [1, 2, 3]
if items:
    print("List has items.")   # Runs — non-empty list is truthy

count = 0
if count:
    print("Non-zero.")
else:
    print("Count is zero.")    # Runs — 0 is falsy
````

## 7. Nested if Statements
Place `if` blocks inside other `if` blocks. Indentation is critical.

````python
is_member = True
balance = 150

if is_member:
    if balance >= 100:
        print("Purchase approved.")    # Runs
    else:
        print("Insufficient balance.")
else:
    print("Members only.")
````

## 8. One-Line if (Ternary Expression)
```python
value = result_if_true  if  condition  else  result_if_false
```

````python
age = 17
status = "adult" if age >= 18 else "minor"
print(status)    # minor
````

## 9. `in` Operator in Conditions
Check membership in strings, lists, tuples, or dicts.

````python
fruits = ["apple", "banana", "cherry"]
if "banana" in fruits:
    print("Banana is available.")    # Runs

role = "admin"
allowed_roles = ["admin", "moderator"]
if role in allowed_roles:
    print("Access granted.")         # Runs
````

## 10. Practical Example — BMI Classifier

````python
def classify_bmi(weight_kg, height_m):
    bmi = weight_kg / (height_m ** 2)
    print(f"BMI: {bmi:.1f}")

    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25.0:
        category = "Normal weight"
    elif bmi < 30.0:
        category = "Overweight"
    else:
        category = "Obese"

    print(f"Category: {category}")

classify_bmi(70, 1.75)    # Normal weight
classify_bmi(50, 1.75)    # Underweight
classify_bmi(100, 1.75)   # Obese
````
